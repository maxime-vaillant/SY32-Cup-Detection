```python
import cv2
import glob
import os
import pandas as pd
from skimage import io, util
from skimage.color import rgb2gray
```

```python
shape_image_x = 400
shape_image_y = 400
```

```python
def generate_positive_train(label_path, image_path, save_path, counter):
    label_files = sorted(glob.glob(os.path.join(label_path, "*.csv")))

    label_df = []

    for label_index, file in enumerate(label_files):
        df = pd.read_csv(file, header=None)
        df['index'] = label_index
        label_df.append(df)

    labels_df = pd.concat(label_df, ignore_index=True)

    for label_index, path in enumerate(label_files):

        if counter > 9999:
            break

        filename = os.path.splitext(path)[0].split('/')[-1]

        img = io.imread('{}{}.jpg'.format(image_path, filename))
        img = util.img_as_float(img)
        img = rgb2gray(img)

        detections = labels_df[labels_df['index'] == label_index].values.tolist()

        for detection_index, detection in enumerate(detections):
            try:

                i, j, h, l, _, _ = [int(x) for x in detection]
                cup = img[i:i+h, j:j+l]
                cup = cv2.resize(cup, (shape_image_x, shape_image_y))


                for rotation in range(4):
                    cup = cv2.rotate(cup, cv2.ROTATE_90_CLOCKWISE)

                    counter += 1

                    if counter > 9999:
                        break

                    io.imsave(save_path+f"{counter:04n}.jpg", cup)

                counter += 1

                if counter > 9999:
                    break

                io.imsave(save_path+f"{counter:04n}.jpg", cup)

            except Exception as err:
                pass

    return counter
```

```python
def generate_negative_train(image_path, save_path, counter):
    negative_images = io.imread_collection(os.path.join(image_path, '*.jpg'))

    for image in negative_images:

        if counter > 9999:
            break

        img = util.img_as_float(image)
        img = rgb2gray(img)

        shape_y, shape_x = img.shape

        if shape_y > shape_image_y*6:
            img = cv2.resize(img, (int(shape_image_y*6 * shape_x / shape_y), shape_image_y*6))
        elif shape_x > shape_image_x*6:
            img = cv2.resize(img, (shape_image_x*6, int(shape_image_x*6 * shape_y / shape_x)))

        shape_y, shape_x = img.shape

        for i in range(shape_y // shape_image_y):
            for j in range(shape_x // shape_image_x):
                neg = img[0+i*shape_image_y:shape_image_y+i*shape_image_y, 0+j*shape_image_x:shape_image_x+j*shape_image_x]

                counter += 1

                if counter > 9999:
                    break

                io.imsave(save_path+f"{counter:04n}.jpg", neg)

    return counter
```

```python
counter = 0

label_path_p22 = 'dataset/train/labels_csv/'
image_path_p22 = 'dataset/train/images/pos/'
label_path_p21 = 'dataset/train_p21/labels_csv/'
image_path_p21 = 'dataset/train_p21/images/pos/'

save_path_pos = 'train_images/pos/'

counter = generate_positive_train(label_path_p22, image_path_p22, save_path_pos, counter)
generate_positive_train(label_path_p21, image_path_p21, save_path_pos, counter)
```

```python
counter = 0

image_path_p22 = 'dataset/train/images/neg/'
image_path_p21 = 'dataset/train_p21/images/neg/'

save_path_neg = 'train_images/neg/'

counter = generate_negative_train(image_path_p22, save_path_neg, counter)
generate_negative_train(label_path_p21, save_path_neg, counter)
```
