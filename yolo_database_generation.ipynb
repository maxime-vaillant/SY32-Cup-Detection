```python
import os
import numpy as np
from lxml import etree
from skimage import io
from sklearn.model_selection import train_test_split
```

```python
!rm -f labels/train/*
!rm -f labels/val/*
```

```python
def generate_images_and_labels_folder(path_label, path_image, path_image_neg):
    labels_csv = os.listdir(path_label)
    neg_images = os.listdir(path_image_neg)

    for lbl in labels_csv:
        root = etree.parse(os.path.join(path_label, lbl))
        objs = root.findall("object")
        bndboxes = np.zeros((len(objs), 5))

        filename = lbl.split('.')[0]
        image = io.imread('{}{}.jpg'.format(path_image, filename))
        shape_y, shape_x, _ = image.shape

        for idx, obj in enumerate(objs):
            bndbox = obj.find("bndbox")

            xmin = int(bndbox.find("xmin").text)
            ymin = int(bndbox.find("ymin").text)
            xmax = int(bndbox.find("xmax").text)
            ymax = int(bndbox.find("ymax").text)

            i = ((xmax + xmin) / 2) / shape_x
            j = ((ymax + ymin) / 2) / shape_y
            w = (xmax - xmin) / shape_x
            h = (ymax - ymin) / shape_y

            bndboxes[idx, :] = np.array([0, i, j, w, h])

        np.savetxt(os.path.join('labels/train', lbl[:-4] + '.txt'), bndboxes, fmt='%f', delimiter=" ")

    for neg in neg_images:
        filename = neg.split('.')[0]

        empty_array = np.array([])

        np.savetxt(os.path.join('labels/train', filename + '.txt'), empty_array)
```

```python
path_label = 'dataset/train/labels'
path_image = 'dataset/train/images/pos/'
path_image_neg = 'dataset/train/images/neg/'

generate_images_and_labels_folder(path_label, path_image, path_image_neg)

path_label_2021 = 'dataset/train_p21/labels'
path_image_2021 = 'dataset/train_p21/images/pos/'
path_image_neg_2021 = 'dataset/train_p21/images/neg/'

generate_images_and_labels_folder(path_label_2021, path_image_2021, path_image_neg_2021)
```
